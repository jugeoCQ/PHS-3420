```python
import pennylane as qml
import numpy
```

```python
# Lancer des tâches sur Yukon 

def ghz():
    """
    Prépare un état GHZ à trois qubits.
    """
    qml.Hadamard(wires = 0)
    qml.CNOT(wires = [0,1])
    qml.CNOT(wires = [1,2])

dev = qml.device("default.qubit", wires = 3)
@qml.set_shots(1000)
@qml.qnode(dev)
def ghz_simulated():
    """Simule la préparation de l’état GHZ et retourne le nombre de bitstrings mesurés.
    """
    ghz()
    return qml.counts()
```
